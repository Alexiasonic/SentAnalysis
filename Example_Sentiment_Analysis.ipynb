```python
import pandas as pd
from textblob import TextBlob

# Sample data: list of economic news headlines
data = {
    'headline': [
        "Economy is booming with new job opportunities",
        "Stock market crashes due to pandemic fears",
        "Central bank announces new stimulus package",
        "Unemployment rates are at an all-time low",
        "Inflation rates are rising rapidly"
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Function to get sentiment
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

# Apply sentiment analysis
df['sentiment'] = df['headline'].apply(get_sentiment)

# Display the DataFrame
print(df)

# Save the DataFrame to visualize later
df.to_csv('sentiment_analysis_output.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Plotting the sentiment scores
plt.figure(figsize=(10, 5))
plt.bar(df.index, df['sentiment'], color='skyblue')
plt.xlabel('News Headline Index')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Analysis of Economic News Headlines')
plt.xticks(df.index, df['headline'], rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Sample data: list of economic news headlines
data = {
    'headline': [
        "Economy is booming with new job opportunities",
        "Stock market crashes due to pandemic fears",
        "Central bank announces new stimulus package",
        "Unemployment rates are at an all-time low",
        "Inflation rates are rising rapidly"
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Function to get sentiment
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

# Apply sentiment analysis
df['sentiment'] = df['headline'].apply(get_sentiment)

# Display the DataFrame
print(df)

# Generate a word cloud
text = " ".join(headline for headline in df.headline)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Economic News Headlines')
plt.show()

# Plotting the sentiment scores
plt.figure(figsize=(10, 5))
plt.bar(df.index, df['sentiment'], color='skyblue')
plt.xlabel('News Headline Index')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Analysis of Economic News Headlines')
plt.xticks(df.index, df['headline'], rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Save the DataFrame to visualize later
df.to_csv('sentiment_analysis_output.csv', index=False)
```
